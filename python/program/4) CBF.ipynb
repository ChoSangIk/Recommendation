### genres column을  movie id별로 나누어준다
* DataFrame원리를 이해해야한다. (아래의 용어는 dict의 dataframe폴더에.)
* Series+
* apply+
* stack()+


```python
test = pd.DataFrame(
    data=movies['genres'].str.split('|') 
    .apply(pd.Series))
test
```

```python
movieGenres = pd.DataFrame(
    data=movies['genres']  
    .str
    .split('|')   
    .apply(pd.Series)
    .stack() # 정리하기 편하게 stack으로 쌓아준다.
    ,columns=['genre'])
movieGenres.index = movieGenres.index.droplevel(1)
movieGenres
```

### genre별로 idf를 구해주자.
* 필요한것.
* 전체갯수, genre별 전체갯수

```python
genres = pd.DataFrame(data = movieGenres.groupby('genre')['genre'].count())
genres.columns= ['movieCount']
totalItems = movies.shape[0]
genres['idf'] = genres['movieCount'].apply(lambda x: math.log10(totalItems/x))
genres
```

### movieGenres에 idf가 나온것을 합쳐준다.

```python
movieGenreWeights = movieGenres.join(genres['idf'], on='genre')
movieGenreWeights
```

### 합쳐서 Weight matrix를 만들것이다.
* movieGenreWeights의 각 영화별 장르,장르별idf와, movie의 movieId를 가져온다.
* 각장르를 column으로 하고, 값을 idf로 한다.

```python
movieWeights = movies[['movieId']]

for genre in genres.index: # genres들을 순서대로 뽑아준다.
    movieWeights = movieWeights\
    .join(movieGenreWeights[movieGenreWeights['genre'] == genre]
    # movieGenreWeights['genre']와 genre가 일치하는것들만 뽑아내서 join시킨다.
    # 이유: 각 genre별 column을 만들어주기위해.
    [['idf']].rename(columns={'idf':genre})) 
    #movieGenreWeights에 있던 idf를 genre와 합쳐준다.
    movieWeights.fillna(0, inplace=True)
    # 빈값인 NaN값을 0 으로 채워준다. 
movieWeights
```

### numpy를 이용해 dataframe형식에서의 영화별 matrix norm을 구해준다.
* iloc은 해당 범위만 읽는것이다.
* value는 밸류값만 가져오는것이다.

```python

movieNorms = pd.DataFrame(
    data=LA.norm(movieWeights
                 .iloc[:,1:] # movieId를 빼고 읽는다.
                 .values, 
                 # value를 통해 값만 가져온다. numpy.array형식으로 바꾸어주는건데
                 #그냥 빠지고 dataframe자체로 넣어도 문제없다.
                ord=2, axis=1),
    #index=movieWeights.index, << 없어도 상관없음
    columns=['norm2'])
movieNorms
```

#### 가중치에  norm을 나눌떄 normalized된다. 왤까?

```python
normalizedMovieWeights = pd.DataFrame(index=movieWeights.index) #빈 데이타 프레임만들기
norms = movieNorms['norm2']
for genre in genres.index:
    normalizedMovieWeights[genre] = movieWeights[genre].divide(norms)
```

```python
movieWeights
```

#### normalized 된것의 CosineSimilarity를 구해준다.
* 요기서 원래 cosine similarity구할때 divide하는데 왜 divide는 안한지 모르겠음 한번 알아보겠음...
* noarmalized된것을 inner product하면 divide안해도 되는걸수도있음...
* ==> 실험삼아서 테스트해본결과 normalized한 값을 matmul하면 Cosine similarity값이 나옴. 
*  밑의 식에선 왜 뒤에 T를 붙였는지 모르겠음.
* => 아마 길이가 긴거에 붙이는거같은데 정확한 수학적인 이유는 모르겠음

```python
movieWeights.columns
```

```python
sims = pd.DataFrame(data=np.matmul(normalizedMovieWeights, normalizedMovieWeights.T), index=movieWeights['movieId'])
sims.columns = index=movieWeights['movieId']
```

 ### 평점을 예측해본다. ( PPT참고)
* 1) recSimSums는 sims에서 유저가 평점을 매겼던 영화전체 영화들과 다른영화간의 유사도를 합한 값이다.
* ex)  유저가 19,88,157 번영화를 봤다면,   19번영화와 1번영화, 88번영화와 1번영화, 157번영화와 1번영화의합이 1번영화의 recSimSums값이다.
* 2) recWeightedRatingSums는 영화끼리의 similarity 데이터프레임에서 user가 Rating한 영화만골라, user가 Ratings한 rating점수를 곱해준다
* ex) 유저가 1번영화를 보고 평점을 매겼다면, 1번과 다른영화들의 유사도들이 있는 sims에서 1번 영화 평점을 다른영화들의 유사도와 곱해준다.저
* ex) 유사도는0.7 이전영화평점은 3점이라면 2.1
* 3) recItemRatings란 영화마다 쌓인 유사도x평점의 합값에 해당 영화 전체 유사도 합을 나눠준다.

* 그러면 결과적으로 전체영화의 rating prediction값이 나오게된다.

```python
recSimSums = sims.loc[userRatings['movieId'].values, :].sum().values
recWeightedRatingSums = np.matmul(sims.loc[userRatings['movieId'].values, :].T.values, userRatings['rating'].values)
recItemRatings = pd.DataFrame(data=np.divide(recWeightedRatingSums, recSimSums), index=sims.index)
recItemRatings.columns = ['prediction']
```

```python
displayMovies(recItemRatings.sort_values(by='prediction', ascending=False).head(30).index)
```

### 정확도를 분석해본다.

```python
userTestRatings = pd.DataFrame(data=test[test['userId'] == userId])
temp = userTestRatings.join(recItemRatings.loc[userTestRatings['movieId']], on='movieId')
mae = (temp['rating'] - temp['prediction']).abs().mean()
rmse = math.sqrt((temp['rating'] - temp['prediction']).pow(2).mean())
print(" MAE:", mae)
print("RMSE:", rmse)
```
